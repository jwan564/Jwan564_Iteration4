```python
finaldataset = finaldataset.withColumn("DistanceFromHome", when((finaldataset["DistanceFromHome"] <= 7), "Close") \
                                       .when((finaldataset["DistanceFromHome"] > 7) & (finaldataset["DistanceFromHome"] <=14), "Medium")\
                                      .when((finaldataset["DistanceFromHome"] > 14) & (finaldataset["DistanceFromHome"] <=21) , "Far")\
                                      .otherwise("Very Far"))
```

```python
finaldataset.select("DistanceFromHome").show(20)
```

```python
from pyspark.ml.classification import LogisticRegression
from pyspark.ml.feature import (VectorAssembler,VectorIndexer, OneHotEncoder,StringIndexer)
```

```python
finaldataset = finaldataset.withColumn("Attrition", when((finaldataset["Attrition"] == "Yes"), 1) \
                                      .otherwise(0))
finaldataset = finaldataset.withColumn("Attrition", finaldataset["Attrition"].cast(T.IntegerType()))
```

```python
train, test = finaldataset.randomSplit([0.7, 0.3])
```

```python
BTravel_indexer = StringIndexer(inputCol='BusinessTravel',outputCol='BTravelIndex')
BTravel_encoder = OneHotEncoder(inputCol='BTravelIndex',outputCol='BTravelVec')
```

```python
Department_indexer = StringIndexer(inputCol='Department',outputCol='DepartmentIndex')
Department_encoder = OneHotEncoder(inputCol='DepartmentIndex',outputCol='DepartmentVec')
```

```python
DFromhome_indexer = StringIndexer(inputCol='DistanceFromHome',outputCol='DFromhomeIndex')
DFromhome_encoder = OneHotEncoder(inputCol='DFromhomeIndex',outputCol='DFromhomeVec')
```

```python
EducationField_indexer = StringIndexer(inputCol='EducationField',outputCol='EducationFieldIndex')
EducationField_encoder = OneHotEncoder(inputCol='EducationFieldIndex',outputCol='EducationFieldVec')
```

```python
Gender_indexer = StringIndexer(inputCol='Gender',outputCol='GenderIndex')
Gender_encoder = OneHotEncoder(inputCol='GenderIndex',outputCol='GenderVec')
```

```python
JobRole_indexer = StringIndexer(inputCol='JobRole',outputCol='JobRoleIndex')
JobRole_encoder = OneHotEncoder(inputCol='JobRoleIndex',outputCol='JobRoleVec')
```

```python
MaritalStatus_indexer = StringIndexer(inputCol='MaritalStatus',outputCol='MaritalStatusIndex')
MaritalStatus_encoder = OneHotEncoder(inputCol='MaritalStatusIndex',outputCol='MaritalStatusVec')
```

```python
assembler = VectorAssembler(inputCols=['BTravelVec',
 'DepartmentVec',
 'DFromhomeVec',
 'EducationFieldVec',
 'GenderVec',
 'JobRoleVec',
 'MaritalStatusVec',
 'EmployeeID','JobInvolvement','PerformanceRating','EnvironmentSatisfaction','JobSatisfaction',
 'WorkLifeBalance','Age','Education','JobLevel','MonthlyIncome','NumCompaniesWorked',
 'PercentSalaryHike','StockOptionLevel','TotalWorkingYears','TrainingTimesLastYear',
 'YearsAtCompany','YearsSinceLastPromotion','YearsWithCurrManager','AvergeWorkingHours'],outputCol='features')
```

```python
from pyspark.ml import Pipeline
```

```python
log_reg = LogisticRegression(featuresCol='features',labelCol='Attrition')
```

```python
pipeline = Pipeline(stages=[BTravel_indexer,Department_indexer,
                           DFromhome_indexer,EducationField_indexer,Gender_indexer,JobRole_indexer,
                            MaritalStatus_indexer,BTravel_encoder,Department_encoder,DFromhome_encoder,
                            EducationField_encoder,Gender_encoder,JobRole_encoder,MaritalStatus_encoder,
                           assembler,log_reg])
```

```python
fit_model = pipeline.fit(train)
```

```python
results = fit_model.transform(test)
```

```python
from pyspark.ml.evaluation import BinaryClassificationEvaluator
```

```python
my_eval = BinaryClassificationEvaluator(rawPredictionCol='prediction',
                                       labelCol='Attrition')
```
